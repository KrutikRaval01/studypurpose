**8.Demonstrate the univariate analysis using distplot and conclude the results (2 Marks)**

```python
# Code Here

import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset (e.g., iris dataset)
iris = sns.load_dataset("iris")

# Plot a univariate distribution (e.g., sepal length)
sns.set(style="whitegrid")  # Set the plot style

# Create a distplot for sepal length
sns.distplot(iris["sepal_length"], kde=True, rug=True, color="blue")

# Add labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")

# Show the plot
plt.show()
```

**16.Draw the pie chart using matplotlib and show case following features  (2 Marks)**
 -  a.labels
 -  b.explode
 -  c.shadow
 -  d.legend

```python
# Code Here

import matplotlib.pyplot as plt
import numpy as np

y = np.array([55, 15, 15, 10,4,1])
mylabels = ["potatoes", "onions", "tomatoes", "garlic","jinger","lemon"]
myexplode = [0.2, 0, 0.1, 0,0,0.3]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```
